```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data=pd.read_csv('/content/insurance_claims.csv')
data.head(2)
```

# Age

```python
#How does age influence the rate of fraud
sns.set_theme(style='darkgrid',palette="rocket")
plt.figure(figsize=(10,6))
sns.boxplot(x='fraud_reported',y='age',data=data, palette='rocket',hue='policy_state')
plt.title("Boxplot of age by policy-state")
```

```python
#How does age influence the claim amount
sns.set_theme(style='darkgrid',palette="viridis")
plt.figure(figsize=(10,6))
sns.relplot(data=data, x="total_claim_amount", y="age",col="insured_sex",kind="scatter",hue="fraud_reported")
```

```python
#How does age influence the rate of fraud
sns.set_theme(style='darkgrid',palette="rocket")
plt.figure(figsize=(10,6))
sns.boxplot(x='incident_type',y='age',data=data, palette='rocket',hue='fraud_reported')
plt.title("Boxplot of age by incident_type")
```

```python
#How does age influence the claim amount
sns.set_theme(style='darkgrid',palette="rocket")
plt.figure(figsize=(10,6))
sns.countplot(x='bodily_injuries',data=data, hue='fraud_reported',palette='rocket')
```

```python
sns.countplot(x='incident_hour_of_the_day',data=data,palette='rocket')
```

```python
#How does age influence the claim amount
sns.set_theme(style='darkgrid',palette="crest")
plt.figure(figsize=(20,10))
sns.relplot(data=data, x="incident_hour_of_the_day", y="age",col="fraud_reported",kind="line")
sns.relplot(data=data, x="incident_hour_of_the_day", y="age",col="insured_sex",kind="line")
```

```python
#How does age influence the claim amount
sns.set_theme(style='darkgrid',palette="rocket")
plt.figure(figsize=(10,6))
#sns.boxplot(data=data,x='incident_severity',y='age',hue='insured_sex')
sns.countplot(x='incident_severity', data=data, hue='fraud_reported' ,palette='rocket')
```
